## Opening file using request Library and assigning it to a varribale

```python
#opening the file using Library and then assigning it to a varriable.

url = 'https://en.wikipedia.org/wiki/List_of_postal_codes_of_Canada:_M'

r = requests.get(url)
content =r.content
```

## Converting HTML format to BeautifulSoup Tree Format

```python

soup = BeautifulSoup(content,'html.parser') 
```

## Finding Relevent data and Assigning them to Lists

```python
A = []
B = []
C = []


for row in soup.findAll('tr'):
    cells = row.findAll('td')  
    
    
    if len(cells) == 3:
        if cells[1].find(text=True).rstrip()=='Not assigned':
            continue
        else :
            A.append(cells[0].find(text=True).rstrip())
            B.append(cells[1].find(text=True).rstrip())
            C.append(cells[2].find(text=True).rstrip())

```

## Assigning Lists into Columns

```python

df = pd.DataFrame(A,columns = ['postal_code'])
df['Brough'] = B
df['Neighborhood'] = C

df
```

## Replacing '/' value with ',' in Neighborhoods Column

```python
df['Neighborhood'] = df['Neighborhood'].str.replace('/' ,',')
df
```

## Importing CSV file with GPS Coordinates

```python
df_Lat_long = pd.read_csv("Long_Lat.csv")
df_Lat_long
```

## Assigning Longitude and Latitude to the Original Dataframe

```python
df["Latitude"] = df_Lat_long["Latitude"]
df["Longitude"] = df_Lat_long["Longitude"]
df
```

## Extracting Data set which is included in Toronto

```python
toronto_data = df[df['Brough'].str.contains('Toronto',case = False)].reset_index(drop=True)
toronto_data.head()
```

## Getting GPS coordinates of Toronto

```python
!conda install -c conda-forge geopy --yes # uncomment this line if you haven't completed the Foursquare API lab
from geopy.geocoders import Nominatim # convert an address into latitude and longitude values

address = 'Toronto, Canada'

geolocator = Nominatim(user_agent="ny_explorer")
location = geolocator.geocode(address)
latitude = location.latitude
longitude = location.longitude
print('The geograpical coordinate of Toronto are {}, {}.'.format(latitude, longitude))
```

## Creating a map of Toronto Using Folium Library

```python
import folium

# create map of Manhattan using latitude and longitude values
map_toronto = folium.Map(location=[latitude, longitude], zoom_start=11)

# add markers to map
for lat, lng, label in zip(toronto_data['Latitude'], toronto_data['Longitude'], toronto_data['Neighborhood']):
    label = folium.Popup(label, parse_html=True)
    folium.CircleMarker(
        [lat, lng],
        radius=5,
        popup=label,
        color='blue',
        fill=True,
        fill_color='#3186cc',
        fill_opacity=0.7,
        parse_html=False).add_to(map_toronto)  
    
map_toronto
```
